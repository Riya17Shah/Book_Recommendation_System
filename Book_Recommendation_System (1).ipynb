# Popularity based Recommendation

```python
rating_with_name = rating.merge(book,on='ISBN')
```

```python
num_rating_df = rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df.head()
```

```python
popularity_df = popularity_df[popularity_df['num_ratings'] >= 250].sort_values('avg_ratings',ascending=False).head(25)
```

```python
popularity_df = popularity_df.merge(book,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

# Collaborative based filtering

```python
rating_with_name.head()
```

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
countable_rating = x[x].index
```

```python
print(x)
```

```python
filter_rating = rating_with_name[rating_with_name['User-ID'].isin(countable_rating)]
```

```python
filter_rating.head()
```

```python
filter_rating.shape
```

```python
y = filter_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_book = y[y].index
```

```python
final_rating = filter_rating[filter_rating['Book-Title'].isin(famous_book)]
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend(book_name):
  index = np.where(pt.index == book_name)[0][0]
  similar_item = sorted(list(enumerate(similarity_score[index])),key =lambda x:x[1],reverse=True)[1:6]

  data = []
  for i in similar_item:
        item = []
        temp_df = book[book['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

  return data



```

```python
recommend('1984')
```
